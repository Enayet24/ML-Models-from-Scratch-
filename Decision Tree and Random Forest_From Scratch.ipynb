*   **Parch:**	number of parents / children aboard the Titanic. The dataset defines family relations in this way:
                 Parent = mother, father
                 Child = daughter, son, stepdaughter, stepson

*   **Fare:**	Passenger fare

*   **Cabin:**	Cabin number

*   **Embarked:**	Port of Embarkation
                 C = Cherbourg
                 Q = Queenstown
                 S = Southampton

*   **Survived:**	Survival Status
                 0 = No
                 1 = Yes

# Implementing Decision Tree and Random Forest from Scratch

```python

class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth

    def gini(self, y):
        classes = np.unique(y)
        impurity = 1
        for cls in classes:
            p = np.sum(y == cls) / len(y)
            impurity -= p ** 2
        return impurity

    def split(self, X_column, threshold):
        left_index = np.where(X_column <= threshold)[0]
        right_index = np.where(X_column > threshold)[0]
        return left_index, right_index

    def find_best_split(self, X, y):
        best_gain = 0
        best_feature, best_threshold = None, None
        parent_gini = self.gini(y)
        n_samples, n_features = X.shape

        for feature in range(n_features):
            X_feature = X[:, feature].astype(float)
            thresholds = np.unique(X_feature)
            for threshold in thresholds:
                left_index, right_index = self.split(X_feature, threshold)
                if len(left_index) == 0 or len(right_index) == 0:
                    continue

                left_gini = self.gini(y[left_index])
                right_gini = self.gini(y[right_index])
                n_left, n_right = len(left_index), len(right_index)
                weighted_gini = (n_left / n_samples) * left_gini + (n_right / n_samples) * right_gini
                gain = parent_gini - weighted_gini

                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature
                    best_threshold = threshold

        return best_feature, best_threshold

    def majority_class(self, y):
        unique_classes = np.unique(y)
        max_count = -1
        majority_cls = None
        for cls in unique_classes:
            count = np.sum(y == cls)
            if count > max_count:
                max_count = count
                majority_cls = cls
        return majority_cls

    def fit(self, X, y, depth=0):
        if len(np.unique(y)) == 1:
            return {"leaf": True, "class": y[0]}
        if self.max_depth is not None and depth >= self.max_depth:
            return {"leaf": True, "class": self.majority_class(y)}

        best_feature, best_threshold = self.find_best_split(X, y)
        if best_feature is None:
            return {"leaf": True, "class": self.majority_class(y)}

        left_index, right_index = self.split(X[:, best_feature], best_threshold)
        left_node = self.fit(X[left_index, :], y[left_index], depth + 1)
        right_node = self.fit(X[right_index, :], y[right_index], depth + 1)

        return {
            "leaf": False,
            "feature": best_feature,
            "threshold": best_threshold,
            "left": left_node,
            "right": right_node
        }

    def predict(self, node, x):
        if node["leaf"]:
            return node["class"]
        if x[node["feature"]] <= node["threshold"]:
            return self.predict(node["left"], x)
        else:
            return self.predict(node["right"], x)


#Creating the model and fitting the training data
decision_tree = DecisionTree(max_depth=5)
final_tree = decision_tree.fit(X_train_np, y_train_np) # returns the information of the final tree in a suitable data structure

# Predictions on the test set
y_pred_dt_scr = np.array([decision_tree.predict(final_tree, instance) for instance in X_test_np])


#calculating the accuracy manually
accuracy_dt_scr = 0


correct_predictions = np.sum(y_pred_dt_scr == y_test_np)

total_datapoints = len(y_test_np)

accuracy_dt_scr = correct_predictions / total_datapoints


print(f"Decision Tree Classifier (from scratch) Accuracy: {round(accuracy_dt_scr,3)}")
```

```python
import numpy as np

class RandomForest:
    def __init__(self, n_estimators=100, max_depth=None):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.trees = []

    def bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        indices = np.random.choice(n_samples, n_samples, replace=True)
        X_sample = X[indices]
        y_sample = y[indices]
        return X_sample, y_sample

    def fit(self, X, y):
        for i in range(self.n_estimators):
            X_sample, y_sample = self.bootstrap_sample(X, y)
            tree_model = DecisionTree(max_depth=self.max_depth)
            fitted_tree = tree_model.fit(X_sample, y_sample)
            self.trees.append((tree_model, fitted_tree))

    def predict_instance(self, x):
        predictions = []
        for tree_model, tree in self.trees:
            pred = tree_model.predict(tree, x)
            predictions.append(pred)
        predictions = np.array(predictions)
        unique_classes, counts = np.unique(predictions, return_counts=True)
        return unique_classes[np.argmax(counts)]

    def predict(self, X):
        y_pred = np.array([self.predict_instance(instance) for instance in X])
        return y_pred


random_forest = RandomForest(n_estimators=25, max_depth=5)
random_forest.fit(X_train_np, y_train_np)


y_pred_rf_scr = random_forest.predict(X_test_np)

accuracy_rf_scr = 0

correct_predictions = np.sum(y_pred_rf_scr == y_test_np)
total_points = len(y_test_np)
accuracy_rf_scr = correct_predictions / total_points

print(f"Random Forest Classifier (from scratch) Accuracy: {round(accuracy_rf_scr,3)}")
```
